```python
df = pd.read_csv('dealer_scrape.csv')
```

```python
df.dtypes
```

```python
df.isnull().sum()
```

```python
df = df.dropna(subset = ['price'])
```

```python
df.isnull().sum()
```

```python
df[df.model.isnull()]
```

```python
missing_models = df[df.model.isnull()].index
```

```python
for item in missing_models:
    print(df.title.loc[item])
```

```python
for item in missing_models:
    df.model.loc[item] = df.title.loc[item].split()[3]
```

```python
df[df.make =='alfaromeo']
```

```python
df.isnull().sum()
```

```python
df.trim = df.trim.fillna('None')
```

```python
df.isnull().sum()
```

```python
df.transmission = df.transmission.fillna('None')
```

```python
df.groupby('condition').count()
```

```python
new_cars = df[df.condition=='New'].index
```

```python
df = df.drop(new_cars)
```

```python
df = df.drop('condition',axis = 1)
```

```python
df.isnull().sum()
```

```python
df.shape
```

```python
df = df.dropna()
```

```python
df.shape
```

```python
df.isnull().sum().sum()
```

```python
df.columns
```

```python
df.price=df.price.str.replace('$','')
df.price=df.price.str.replace(',','')
df.price = df.price.apply(lambda x: x.split('.')[0])
df.price = df.price.astype('int')
df.price.head()
```

```python
df.year = df.year.astype('int')
```

```python
df.bodytype.unique()
```

```python
df['bodytype'] = df['bodytype'].replace(['SUV, Crossover', 'Pickup Truck', 'Coupe (2 door)',
       'Minivan, Van'],['SUV','Truck','Coupe','Minivan'])
df.bodytype.unique()
```

```python
df.transmission.unique()
```

```python
df['transmission'] = df['transmission'].replace(['None','0','Other'], 'Automatic')
df.transmission.value_counts()
```

```python
df.odometer.dtype
```

```python
df.odometer = df.odometer.str.replace(',','')
df.odometer = df.odometer.astype('int')
```

```python
df.drivetrain.unique()
```

```python
df['drivetrain'] = df['drivetrain'].replace(['Front-wheel drive (FWD)','4 x 4',
       'All-wheel drive (AWD)', 'Rear-wheel drive (RWD)'],['FWD','4x4','AWD','RWD'])
```

```python
df.to_csv('ready_for_eda.csv',index = False)
```
